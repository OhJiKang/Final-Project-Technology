```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data points
x = np.array([100, 200, 300, 400,500,600,1000,2000,24000]).reshape(-1, 1)
y = np.array([1.73,3.06,4.38,6.01,7.04,8.12,12.59,23.26,78])

# Polynomial regression
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_pred = model.predict(x_poly)

# New test data
x_test = np.array([3000,4000,5000,10000,20000,30000,40000,50000]).reshape(-1, 1)
x_test_poly = poly.transform(x_test)
y_test_pred = model.predict(x_test_poly)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Actual data')
plt.plot(x, y_pred, color='blue', label='Polynomial fit (degree=3)')
plt.scatter(x_test, y_test_pred, color='green', label='Predictions for x_test', marker='x')
# plt.scatter(500, 162, color='black', label='Actual Value for 500 products')
plt.xlabel('Number of products')
plt.ylabel('Time (s)')
plt.title('Time vs Number of Products')
plt.legend()
plt.grid(True)
plt.show()
```
